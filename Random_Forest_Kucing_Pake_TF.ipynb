```python
import tensorflow_decision_forests as tfdf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from google.colab import files
```

```python
uploaded = files.upload()
data = pd.read_csv("Dummy.csv")
print(data.head())
```

```python
# Buat Cek Aja
print(data.columns)
```

```python
gejala_columns = data.columns[1:-1]
label_column = data.columns[-1]
```

```python
features = data[gejala_columns]
labels = data[label_column]
```

```python
test_size = 0.15
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, test_size=test_size, random_state=122
)
```

```python
train_features[label_column] = train_labels
test_features[label_column] = test_labels
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_features, label=label_column)
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_features, label=label_column)
```

```python
#Random Forest model pakai TF
model = tfdf.keras.RandomForestModel(num_trees=202, max_depth=20)
model.fit(train_ds)

model.summary()

# Evaluate  model.
model.evaluate(test_ds)
print(model.evaluate(test_ds))
```

```python
tfdf.model_plotter.plot_model_in_colab(model, tree_idx=15)
```

```python
model_json = model.to_json()

# Save the model to a JSON file
with open("model.json", "w") as json_file:
    json_file.write(model_json)

# Download the JSON file
files.download('model.json')

# Save the model weights
model.save_weights('model_weights.h5')

# Download the model weights
files.download('model_weights.h5')
```
